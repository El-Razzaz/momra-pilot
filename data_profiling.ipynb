```python
dataset_df = closed_cases_df[closed_cases_df["مصدر البلاغ"].str.contains('تطبيق بلدي',na=False)][["الوصف",'التصنيف التخصصي']]
```

```python
dataset_df = dataset_df.dropna()
```

```python
train_df =  dataset_df.groupby(['التصنيف التخصصي'], group_keys=False).apply(lambda x: x.sample(frac=.6)).sample(frac=1)
test_df =  dataset_df.drop(train_df.index).groupby('التصنيف التخصصي', group_keys=False).apply(lambda x: x.sample(frac=0.5)).sample(frac=1)
validation_df = dataset_df.drop(train_df.index).drop(test_df.index).groupby('التصنيف التخصصي', group_keys=False).apply(lambda x: x.sample(frac=1)).sample(frac=1)
```

```python
x_test = test_df['الوصف']
y_test = test_df['التصنيف التخصصي']
x_train = train_df['الوصف']
y_train = train_df['التصنيف التخصصي']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# Vectorize text reviews to numbers
vec = CountVectorizer()
x = vec.fit_transform(x_train).toarray()
x_test = vec.transform(x_test).toarray()
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(x, y_train)
model.score(x_test, y_test)
```
